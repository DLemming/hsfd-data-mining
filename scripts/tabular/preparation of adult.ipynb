# Load dataset

```python
import pandas as pd
import numpy as np
from ucimlrepo import fetch_ucirepo
#dataset = pd.read_csv('../datasets/adult_data.csv')
adult = fetch_ucirepo(id=2)
X: pd.DataFrame = adult.data.features
y: pd.DataFrame = adult.data.targets

dataset = X.join(y)

print(dataset.columns)
print(len(dataset))
```

```python
#drop nan values
dataset = dataset.replace('?', np.nan)
dataset = dataset.dropna()
dataset = dataset.drop(['education'], axis=1)
dataset = dataset.rename(columns={"income": "class"})

print(len(dataset))
```

### Define train and test and save them. We are going to use always the same for reproducibility reasons.


```python
#test set 30
from sklearn.model_selection import train_test_split
label = dataset.pop('class')
train_set, test_set, train_label, test_label = train_test_split(dataset,label,test_size=0.30, random_state=42)
```

```python
import pickle
nome = 'adult'
title = "../datasets/train_set_"+nome+"_strat.p"
with open(title, 'wb') as fp:
    pickle.dump(train_set, fp)
title = "../datasets/train_label_"+nome+"_strat.p"
with open(title, 'wb') as fp:
    pickle.dump(train_label, fp)
title = "../datasets/test_set_"+nome+"_strat.p"
with open(title, 'wb') as fp:
    pickle.dump(test_set, fp)
title = "../datasets/test_label_"+nome+"_strat.p"
with open(title, 'wb') as fp:
    pickle.dump(test_label, fp)

```
